```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
import pandas
import sys, os

NewDataPath = "/work/mk423/Anxiety/New_FLX_Animals_April_12.pkl"
OldDataPath = "/work/mk423/Anxiety/FLX_{}_dict_old_features.pkl"

FEATURE_LIST = ["X_psd","X_coh","X_gc"]
FEATURE_WEIGHT = [10,1,1]
```

```python
448/56
```

```python
newDict = pickle.load(open(NewDataPath,"rb"))
oldTrainDict = pickle.load(open(OldDataPath.format("train"),"rb"))
oldValDict = pickle.load(open(OldDataPath.format("validation"),"rb"))
oldTestDict = pickle.load(open(OldDataPath.format("test"),"rb"))
```

```python
plt.figure(figsize=(20,20))
for i in range(8):
    plt.subplot(4,2,i+1)
    X_new = newDict["X_psd"][newDict["y_hab"].squeeze()==1]
    plt.plot(np.mean(X_new[:,i*56:(i+1)*56],axis=0),label='newholdout')
    plt.fill_between(np.arange(56),
                     np.mean(X_new[:,i*56:(i+1)*56],axis=0)+np.std(X_new[:,i*56:(i+1)*56],axis=0),
                     np.mean(X_new[:,i*56:(i+1)*56],axis=0)-np.std(X_new[:,i*56:(i+1)*56],axis=0),
                     color="blue",
                    alpha=0.1)
    plt.plot(np.mean(oldTrainDict["X_psd"][:,i*56:(i+1)*56],axis=0),label="train")
    plt.fill_between(np.arange(56),
                     np.mean(oldTrainDict["X_power_1_2"][:,i*56:(i+1)*56],axis=0)+np.std(oldTrainDict["X_power_1_2"][:,i*56:(i+1)*56],axis=0),
                     np.mean(oldTrainDict["X_power_1_2"][:,i*56:(i+1)*56],axis=0)-np.std(oldTrainDict["X_power_1_2"][:,i*56:(i+1)*56],axis=0),
                     color="orange",
                    alpha=0.1)
    plt.plot(np.mean(oldValDict["X_psd"][:,i*56:(i+1)*56],axis=0),label="val")
    plt.fill_between(np.arange(56),
                     np.mean(oldValDict["X_power_1_2"][:,i*56:(i+1)*56],axis=0)+np.std(oldValDict["X_power_1_2"][:,i*56:(i+1)*56],axis=0),
                     np.mean(oldValDict["X_power_1_2"][:,i*56:(i+1)*56],axis=0)-np.std(oldValDict["X_power_1_2"][:,i*56:(i+1)*56],axis=0),
                     color="green",
                    alpha=0.1)
    plt.plot(np.mean(oldTestDict["X_psd"][:,i*56:(i+1)*56],axis=0),label="test")
    plt.fill_between(np.arange(56),
                     np.mean(oldTestDict["X_power_1_2"][:,i*56:(i+1)*56],axis=0)+np.std(oldTestDict["X_power_1_2"][:,i*56:(i+1)*56],axis=0),
                     np.mean(oldTestDict["X_power_1_2"][:,i*56:(i+1)*56],axis=0)-np.std(oldTestDict["X_power_1_2"][:,i*56:(i+1)*56],axis=0),
                     color="red",
                    alpha=0.1)
    plt.title("Power Feature: {}".format(i))
    plt.legend()
plt.suptitle("Feature Distribution Comparisons")
plt.show()
```
