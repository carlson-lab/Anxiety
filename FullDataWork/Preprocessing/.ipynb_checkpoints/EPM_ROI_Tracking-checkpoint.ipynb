```python
import numpy as np
from scipy.io import loadmat
import sys, os

refDir = "/work/mk423/EPM_lfps/Full_ClassData/"
```

```python
files = os.listdir(refDir)
files = [file for file in files if file.split('.')[-1]=="mat"]
```

```python
from scipy.stats import mode

def mean_consecutive_ones(binary_vector):
    consecutive_counts = []
    current_count = 0

    for bit in binary_vector:
        if bit == 1:
            current_count += 1
        else:
            if current_count > 0:
                consecutive_counts.append(current_count)
                current_count = 0

    # Append the last count if the sequence ends with 1's
    if current_count > 0:
        consecutive_counts.append(current_count)

    # Calculate the mean
    mean_ones = sum(consecutive_counts) / len(consecutive_counts) if consecutive_counts else 0

    return mean_ones


    
holdout_mice = ["Mouse0641",
"Mouse39132",
"Mouse39115",
"Mouse39121",
"Mouse39122",
"Mouse39135",
"Mouse6674",
"Mouse69061",
"Mouse69071",
"Mouse69075",
"Mouse8893"]

n_minutes = 5
max_frames = 50*60*n_minutes

holdout_mice_list = []
n_oa_list = []
holdout_oa_perc = []
holdout_oa_mean_time = []
holdout_ca_perc = []
holdout_ca_mean_time = []
holdout_cen_perc = []
holdout_cen_mean_time = []
holdout_mean_consecutive_oa = []
holdout_mean_consecutive_ca = []
holdout_mean_consecutive_cen = []

for file in files:
    mouse_name = file.split('_')[0]
    
    
    temp_dict = loadmat(refDir + file)
    startFrame = int(temp_dict["StartTime"].astype(float)*50)
    rois = temp_dict["ROI"].squeeze()[startFrame:startFrame+max_frames]
    #print(temp_dict["ROI"].shape[0]//50)
    num_secs_leftover = rois.shape[0]%50
    if num_secs_leftover != 0:
        rois = rois[:-num_secs_leftover].reshape(-1,50)
    else:
        rois = rois.reshape(-1,50)
        
    rois = mode(rois,axis=1)[0].squeeze()
    #print(mouse_name,rois.shape)
    
    ca_labels = np.logical_or(rois==1,
                              rois==3)
    oa_labels = np.logical_or(rois==2,
                              rois==4)
    center_labels = rois==5
    
    n_ca = np.sum(ca_labels)
    n_oa = np.sum(oa_labels)
    n_center = np.sum(center_labels)
    
    mean_cas = mean_consecutive_ones(ca_labels)
    mean_oas = mean_consecutive_ones(oa_labels)
    mean_cens = mean_consecutive_ones(center_labels)
    
    mean_ca_time = np.mean(np.where(ca_labels==1)[0] + 1)
    
    if n_oa > 0:
        mean_oa_time = np.mean(np.where(oa_labels==1)[0] + 1)
    else:
        mean_oa_time = np.nan
        
    mean_cen_time = np.mean(np.where(center_labels==1)[0] + 1)

    perc_ca = n_ca / (rois.shape[0])
    perc_oa = n_oa / (rois.shape[0])#max_frames
    perc_cen = n_center / (rois.shape[0])#max_frames
    
    
    ps = "{} n_secs: {} \npercOA: {:.2f}, meanOAtime: {} \
    \npercCA: {:.2f}, meanCAtime: {} \
    \npercCen: {:.2f}, meanCentime: {} \
    \ntotalPercent: {:.2f}"

    if mouse_name in holdout_mice:

        holdout_mice_list.append(mouse_name)
        n_oa_list.append(n_oa)
        holdout_oa_perc.append(perc_oa)
        holdout_oa_mean_time.append(mean_oa_time)
        holdout_ca_perc.append(perc_ca)
        holdout_ca_mean_time.append(mean_ca_time)
        holdout_cen_perc.append(perc_cen)
        holdout_cen_mean_time.append(mean_cen_time)
        holdout_mean_consecutive_ca.append(mean_cas)
        holdout_mean_consecutive_oa.append(mean_oas)
        holdout_mean_consecutive_cen.append(mean_cens)
        #print(rois.shape)
        print(ps.format(mouse_name,
                        rois.shape[0],
                    perc_oa*100,
                    mean_oa_time,
                    perc_ca*100,
                    mean_ca_time,
                    perc_cen*100,
                    mean_cen_time,
                    100*(perc_oa+perc_ca+perc_cen))
             )
        

    
```

```python
import pandas as pd

df = pd.DataFrame.from_dict({
    "mouse":holdout_mice_list,
    "mean consecutive ca windows":holdout_mean_consecutive_ca,
    "mean consecutive oa windows":holdout_mean_consecutive_oa,
    "mean consecutive center windows":holdout_mean_consecutive_cen
})
df.to_csv("/hpc/home/mk423/Anxiety/FullDataWork/Projections/EPM_avg_roi_seg_length.csv")
df
```

```python
import pandas as pd

df_behavior = pd.DataFrame.from_dict({
    "mouse":holdout_mice_list,
    "n_oa_seconds":n_oa_list,
    "oa_perc":holdout_oa_perc,
    "oa_mean_time":holdout_oa_mean_time,
    "ca_perc":holdout_ca_perc,
    "ca_mean_time":holdout_ca_mean_time,
    "center_perc":holdout_cen_perc,
    "center_mean_time":holdout_cen_mean_time,
})

df_behavior.to_csv("/hpc/home/mk423/Anxiety/FullDataWork/Projections/HoldoutMice_EPM_ROI_perc_mean_time_{}_minutes.csv".format(n_minutes)
                   ,index=False)
```
