```python
import numpy as np
import matplotlib.pyplot as plt
import sys, os
from scipy.io import loadmat

sys.path.append("/work/mk423/lpne-data-analysis/")
import data_tools

frame_labels_path = "/work/mk423/OFT/ClassData/"
```

```python
files = os.listdir(frame_labels_path)
files = [file for file in files if file.split('.')[-1]=="mat"]
```

```python
temp_dict = loadmat(frame_labels_path + files[0])
temp_dict.keys()
```

```python
def mean_consecutive_elements(binary_vector):
    consecutive_counts_zeros = []
    consecutive_counts_ones = []
    current_count_zeros = 0
    current_count_ones = 0

    for bit in binary_vector:
        if bit == 0:
            current_count_zeros += 1
            if current_count_ones > 0:
                consecutive_counts_ones.append(current_count_ones)
                current_count_ones = 0
        else:
            current_count_ones += 1
            if current_count_zeros > 0:
                consecutive_counts_zeros.append(current_count_zeros)
                current_count_zeros = 0

    # Append the last counts
    if current_count_zeros > 0:
        consecutive_counts_zeros.append(current_count_zeros)
    if current_count_ones > 0:
        consecutive_counts_ones.append(current_count_ones)

    # Calculate the mean
    mean_zeros = sum(consecutive_counts_zeros) / len(consecutive_counts_zeros) if consecutive_counts_zeros else 0
    mean_ones = sum(consecutive_counts_ones) / len(consecutive_counts_ones) if consecutive_counts_ones else 0

    return mean_zeros, mean_ones
holdout_mice = ["Mouse04203",
"Mouse39115",
"Mouse39121",
"Mouse39122",
"Mouse39132",
"Mouse39135",
"Mouse69061",
"Mouse69071",
"Mouse69075",]


n_minutes = 5
max_frames = 50*60*n_minutes
max_secs = 300
holdout_mice_list = []
holdout_surr_perc = []
holdout_center_perc = []
holdout_surr_mean_times = []
holdout_center_mean_times = []
holdout_mean_surr_seg = []
holdout_mean_ctr_seg = []

for file in files:
    mouse_name = file.split('_')[0]
    
    
    temp_dict = loadmat(frame_labels_path + file)
    startFrame = int(temp_dict["StartTime"].astype(float)*50)
    startSec = int(temp_dict["StartTime"])
    rois = temp_dict["ROI_sec"].squeeze()[startSec:startSec+max_secs]
    #print(temp_dict["ROI"].shape[0]//50)
    
    n_surr = np.sum(rois==0)
    n_center = np.sum(rois==1)
    

    surr_times = np.where(rois==0)[0] + 1
    surr_times = [time for time in surr_times if time <= 300]
    center_times = np.where(rois==1)[0] + 1
    cener_times = [time for time in center_times if time <= 300]
    
    mean_surr_time = np.mean(surr_times)
    mean_center_time = np.mean(center_times)
    
    perc_surr = n_surr / (rois[~np.isnan(rois)].shape[0])
    perc_center = n_center / (rois[~np.isnan(rois)].shape[0])#max_frames
    
    #mean_surr_seg = mean_consecutive_elements(rois==0)
    mean_surr_seg, mean_ctr_seg = mean_consecutive_elements(rois)
    
    ps = "{} \nperc_surr: {:.2f}, surr_time: {:.2f}, \
    \nperc_center: {:.2f} center_time: {:.2f}\n"

    if mouse_name in holdout_mice:
        #print(np.unique(temp_dict["ROI_sec"],return_counts=True),temp_dict["ROI_sec"].shape[0])
        holdout_mice.remove(mouse_name)
        holdout_mice_list.append(mouse_name)
        holdout_surr_perc.append(perc_surr)
        holdout_center_perc.append(perc_center)
        holdout_center_mean_times.append(mean_center_time)
        holdout_surr_mean_times.append(mean_surr_time)
        holdout_mean_surr_seg.append(mean_surr_seg)
        holdout_mean_ctr_seg.append(mean_ctr_seg)
        #holdout_sum_perc.append(perc_surr + perc_center)
        
        print(ps.format(mouse_name,
                    perc_surr*100,
                        mean_surr_time,
                        
                    perc_center*100,
                       mean_center_time)
             )
    
print("Missing: ",holdout_mice)
```
