```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import os

network_array = np.arange(3,61,3)

results_path = "/hpc/home/mk423/Anxiety/FullDataWork/Validations/SingleTask/"
results_file = results_path + "{}_{}_net_val_results.pkl"
figures_path = "/hpc/home/mk423/Anxiety/FullDataWork/Figures/SingleTask/"
def mean_auc(auc_dict):
    temp  = []
    
    for key in list(auc_dict.keys()):
        if not np.isnan(auc_dict[key][0]):
            temp.append(auc_dict[key])
            
    return np.mean(temp)
```

### EPM

```python
files = os.listdir(results_path)

flx_auc_list = []
epm_auc_list = []
oft_auc_list = []

flx_mse_list = []
epm_mse_list = []
oft_mse_list = []

n_nets_list = []
for file in files:
    split_file = file.split('_')
    if split_file[0] == "epm" and split_file[2] == "net":
        
        resultsDict = pickle.load(open(results_path + file,"rb"))
        
        flx_auc_list.append(mean_auc(resultsDict["flx_auc"]))
        epm_auc_list.append(mean_auc(resultsDict["epm_auc"]))
        oft_auc_list.append(mean_auc(resultsDict["oft_auc"]))

        flx_mse_list.append(resultsDict["flx_mse"])
        epm_mse_list.append(resultsDict["epm_mse"])
        oft_mse_list.append(resultsDict["oft_mse"])

        n_nets_list.append(resultsDict["n_nets"])
        

flx_auc_list = np.array(flx_auc_list)
epm_auc_list = np.array(epm_auc_list)
oft_auc_list = np.array(oft_auc_list)

flx_mse_list = np.array(flx_mse_list)
epm_mse_list = np.array(epm_mse_list)
oft_mse_list = np.array(oft_mse_list)

n_nets_list = np.array(n_nets_list)

sort_idxs = np.argsort(n_nets_list)

plt.figure(figsize=(10,5))
plt.plot(n_nets_list[sort_idxs],flx_auc_list[sort_idxs],"-o",label="flx")
plt.plot(n_nets_list[sort_idxs],epm_auc_list[sort_idxs],"-o",label="epm")
plt.plot(n_nets_list[sort_idxs],oft_auc_list[sort_idxs],"-o",label="oft")
plt.xlabel("Number of Components")
plt.ylabel("Mean AUC")
plt.title("Train on EPM Prediction")
plt.legend()
plt.ylim([0.03,1.05])
plt.savefig(figures_path + "EPM_single_task_prediction.svg")

plt.show()

plt.figure(figsize=(10,5))
plt.plot(n_nets_list[sort_idxs],flx_mse_list[sort_idxs],"-o",label="flx")
plt.plot(n_nets_list[sort_idxs],epm_mse_list[sort_idxs],"-o",label="epm")
plt.plot(n_nets_list[sort_idxs],oft_mse_list[sort_idxs],"-o",label="oft")
plt.xlabel("Number of Components")
plt.ylabel("MSE")
plt.title("Train on EPM Reconstruction")
plt.ylim([0.03,0.05])
plt.legend()
plt.savefig(figures_path + "EPM_single_task_reconstruction.svg")
plt.show()
```
