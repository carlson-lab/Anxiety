```python
import numpy as np
import pickle
import matplotlib as mpl
import matplotlib.pyplot as plt
import sys, os

results_path = "/hpc/home/mk423/Anxiety/FullDataWork/Validations/dCSFA_SingleFeature/"
results_file = results_path + "{}_singleFeature_model_30_component_results.pkl"
saveFigLocation = "/hpc/home/mk423/Anxiety/FullDataWork/Figures/"
```

```python
powerFiles = []
cohFiles = []
gcFiles = []

for file in os.listdir(results_path):
    
    if len(file.split("->")) > 1:
        gcFiles.append(file)
    elif len(file.split("-")) > 1:
        cohFiles.append(file)
    else:
        powerFiles.append(file)
        
powerFiles.remove('.ipynb_checkpoints')
```

```python
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42

offset = 0.05
ticks_list = ["dCSFA-NMF"]
mpl.rc("font",size=18)

plt.figure(figsize=(20,10))

plt.scatter(0-offset,0.59,color="blue",label="FLX")
plt.scatter(0,0.76,color="orange",label="EPM")
plt.scatter(0+offset,0.84,color="green",label="OFT")

plt.errorbar(0-offset,0.59,yerr=0.04,color="blue")
plt.errorbar(0,0.76,yerr=0.03,color="orange")
plt.errorbar(0+offset,0.84,yerr=0.03,color="green")

plt.axvline(0.5,0,1,alpha=0.4,linestyle="dashed",color="grey")

plt.axhline(0.59,-1,len(powerFiles)+1,alpha=0.5,linestyle="dashed",color="blue")
plt.axhline(0.76,-1,len(powerFiles)+1,alpha=0.5,linestyle="dashed",color="orange")
plt.axhline(0.84,-1,len(powerFiles)+1,alpha=0.5,linestyle="dashed",color="green")

for idx, file in enumerate(np.sort(powerFiles)):
    
    idx = idx+1
    rDict = pickle.load(open(results_path + file,"rb"))
    
    flx_mean = rDict["flx_mean_test_auc"]
    flx_sterr = rDict["flx_stderr_test_auc"]
    
    epm_mean = rDict["epm_mean_test_auc"]
    epm_sterr = rDict["epm_stderr_test_auc"]
    
    oft_mean = rDict["oft_mean_test_auc"]
    oft_sterr = rDict["oft_stderr_test_auc"]
    
    feature = rDict["feature"]
    ticks_list.append(feature)
    
    plt.scatter(idx-offset,flx_mean,color="blue")
    plt.scatter(idx,epm_mean,color="orange")
    plt.scatter(idx+offset,oft_mean,color="green")
    
    plt.errorbar(idx-offset,flx_mean,yerr=flx_sterr,color="blue")
    plt.errorbar(idx,epm_mean,yerr=epm_sterr,color="orange")
    plt.errorbar(idx+offset,oft_mean,yerr=oft_sterr,color="green")
    
plt.axhline(0.5,-1,len(powerFiles)+1,linestyle="dashdot",color="black")
plt.xticks(range(len(ticks_list)),ticks_list)
plt.legend(loc="lower left")
#plt.ylim([0.3,1])
plt.ylabel("AUC")
plt.title("Single Region Power dCSFA vs Network Model Prediction")
plt.savefig(saveFigLocation + "power_single_feature_dCSFA.svg")
plt.show()
```
