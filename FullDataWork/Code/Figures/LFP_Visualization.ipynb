```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import numpy as np
filepath = "/work/mk423/Anxiety/Mouse0630_110717_EPM_LFP.mat"
plots_path = "/hpc/home/mk423/Anxiety/FullDataWork/Figures/"

dataDict = loadmat(filepath)
```

```python
brain_regions = ['Amy_BLA_01', 'Cg_Cx_L_02', 'Hipp_01',
                 'IL_Cx_L_01', 'MD_Thal_01', 'NAc_Core_01',
                 'PrL_Cx_L_01', 'VTA_L_01']
br_label_list = ["VTA","PrL","NAc","Thal","IL","Hipp","Cg","Amy"]

plt.figure(figsize=(20,5))
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)
for i, key in enumerate(brain_regions):
    bias = 8-i
    data = MinMaxScaler().fit_transform(dataDict[key][0,:6000].reshape(-1,1)).squeeze()
    
    plt.plot(data+bias,color="black")
     
for i in range(0,7000,1000):
    if i!=0 and i!=1000 and i!=6000:
        plt.axvline(i,linestyle="dashed",linewidth=5,alpha=0.5,color="red")
    
plt.yticks(np.arange(8)+1.5,br_label_list)
plt.yticks(fontsize=20)

plt.xticks(np.arange(500,7500,1000),['$v_{t-3}$','$v_{t-2}$','$v_{t-1}$','$v_{t}$',
                                   '$v_{t+1}$','$v_{t+2}$','$v_{t+3}$'])
plt.xticks(fontsize=30)
plt.xticks([])
plt.xlim([1000,6000])
plt.savefig(plots_path + "lfp_example_figure.svg",format="svg")
plt.show()
```
