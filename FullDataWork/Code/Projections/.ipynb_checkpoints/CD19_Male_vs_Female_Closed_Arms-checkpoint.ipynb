```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm 
from statsmodels.formula.api import ols
import statsmodels.formula.api as smf
import statsmodels.api as sm

projections = "/hpc/home/mk423/Anxiety/FullDataWork/Projections/net_{}_scores_headstage_tracked_CD19EPM_closed_arm_scores_over_time_w_HC_mean.csv"
mouse_sexes = "/hpc/home/mk423/Anxiety/FullDataWork/Preprocessing/CLOCK_mice_sexes.xlsx"

def reshape_df(df,data_idxs,group):
    mice = df.mouse
    new_structure = np.zeros((len(data_idxs)*len(mice),4))*np.nan
    for idx, mouse in enumerate(mice):
        new_structure[(len(data_idxs)*idx):(len(data_idxs)*(idx+1)),0] = df[data_idxs].values[idx,:].copy()
        new_structure[(len(data_idxs)*idx):(len(data_idxs)*(idx+1)),1] = idx + 1
        new_structure[(len(data_idxs)*idx):(len(data_idxs)*(idx+1)),2] = data_idxs
        new_structure[(len(data_idxs)*idx):(len(data_idxs)*(idx+1)),3] = group
    return pd.DataFrame(new_structure,columns=["score","mouse","time","geno"])
```

```python
df_sex = pd.read_excel(mouse_sexes,header=None)
```

```python
female_mice = df_sex[df_sex[2]=="F"][0].values
male_mice = df_sex[df_sex[2]=="M"][0].values

clock_female_mice = df_sex[np.logical_and(df_sex[2]=="F",
                                          df_sex[1]=="CLOCK")][0].values
clock_male_mice = df_sex[np.logical_and(df_sex[2]=="M",
                                        df_sex[1]=="CLOCK")][0].values

wt_female_mice = df_sex[np.logical_and(df_sex[2]=="F",
                                          df_sex[1]=="WT")][0].values
wt_male_mice = df_sex[np.logical_and(df_sex[2]=="M",
                                        df_sex[1]=="WT")][0].values
```

```python
def gad(df,idxs,num_secs=10):
    return np.nanmean(df[idxs].values.reshape(df.shape[0],-1,num_secs),axis=2)

networks = [1,2]
all_idxs = np.arange(1,301).astype(str)

for net in networks:
    
    print("\n\n-------NETWORK {}-------\n\n".format(net))
    ca_data = pd.read_csv(projections.format(net))
    
    ca_male = pd.concat([ca_data[ca_data.mouse==mouse] for mouse in male_mice])
    ca_female = pd.concat([ca_data[ca_data.mouse==mouse] for mouse in female_mice])
    
    clock_ca_male = pd.concat([ca_data[ca_data.mouse==mouse] for mouse in clock_male_mice])
    clock_ca_female = pd.concat([ca_data[ca_data.mouse==mouse] for mouse in clock_female_mice])
    
    wt_ca_male = pd.concat([ca_data[ca_data.mouse==mouse] for mouse in wt_male_mice])
    wt_ca_female = pd.concat([ca_data[ca_data.mouse==mouse] for mouse in wt_female_mice])
    
    plt.figure(figsize=(20,5))
    
    plt.subplot(1,2,1)
    
    male_clock_minute_averaged = gad(clock_ca_male,all_idxs)
    male_wt_minute_averaged = gad(wt_ca_male,all_idxs)
    plt.fill_between(np.arange(1,male_clock_minute_averaged.shape[1]+1),
                 np.nanmean(male_clock_minute_averaged,axis=0) + np.nanstd(male_clock_minute_averaged,axis=0)/np.sqrt(clock_male_mice.shape[0]),
                 np.nanmean(male_clock_minute_averaged,axis=0) - np.nanstd(male_clock_minute_averaged,axis=0)/np.sqrt(clock_male_mice.shape[0]),
                 color="grey",
                 alpha=0.4,
                 label="CLOCK Male Mice"
                )
    
    plt.fill_between(np.arange(1,male_wt_minute_averaged.shape[1]+1),
                 np.nanmean(male_wt_minute_averaged,axis=0) + np.nanstd(male_wt_minute_averaged,axis=0)/np.sqrt(wt_male_mice.shape[0]),
                 np.nanmean(male_wt_minute_averaged,axis=0) - np.nanstd(male_wt_minute_averaged,axis=0)/np.sqrt(wt_male_mice.shape[0]),
                 color="navy",
                 alpha=0.4,
                 label="Balb/c Male Mice"
                )
    plt.legend()
    #plt.xticks(np.arange(0,clock_minute_averaged.shape[1],10), np.arange(0,clock_minute_averaged.shape[1],10)*10)
    #plt.ylim([0,0.2])
    plt.title("Network {} Male Mice (10 second average)".format(net))
    
    plt.subplot(1,2,2)
    
    clock_minute_averaged = gad(clock_ca_female,all_idxs)
    wt_minute_averaged = gad(wt_ca_female,all_idxs)
    
    plt.fill_between(np.arange(1,clock_minute_averaged.shape[1]+1),
                 np.nanmean(clock_minute_averaged,axis=0) + np.nanstd(clock_minute_averaged,axis=0)/np.sqrt(clock_male_mice.shape[0]),
                 np.nanmean(clock_minute_averaged,axis=0) - np.nanstd(clock_minute_averaged,axis=0)/np.sqrt(clock_male_mice.shape[0]),
                 color="grey",
                 alpha=0.4,
                 label="CLOCK Female Mice"
                )
    
    plt.fill_between(np.arange(1,wt_minute_averaged.shape[1]+1),
                 np.nanmean(wt_minute_averaged,axis=0) + np.nanstd(wt_minute_averaged,axis=0)/np.sqrt(wt_male_mice.shape[0]),
                 np.nanmean(wt_minute_averaged,axis=0) - np.nanstd(wt_minute_averaged,axis=0)/np.sqrt(wt_male_mice.shape[0]),
                 color="navy",
                 alpha=0.4,
                 label="Balb/c Female Mice"
                )
    plt.legend()
    

    plt.title("Network {} Female Mice (10 second average)".format(net))
    #plt.ylim([0,0.2])
    
    plt.show()
    
    clock_male_stats_df = reshape_df(clock_ca_male,all_idxs,0)
    wt_male_stats_df = reshape_df(wt_ca_male,all_idxs,1)
    
    clock_female_stats_df = reshape_df(clock_ca_female,all_idxs,0)
    wt_female_stats_df = reshape_df(wt_ca_female,all_idxs,1)
    
    male_stats_df = pd.concat([clock_male_stats_df,wt_male_stats_df])
    female_stats_df = pd.concat([clock_female_stats_df,wt_female_stats_df])
    
    formula = "score ~ time + geno"
    
    #Male Model
    model = ols(formula, data=male_stats_df).fit()
    print("\n\nMALE MICE NETWORK {} ANCOVA ".format(net) + formula + "\n\n")
    print(model.summary())
    
    #Female Model
    model = ols(formula, data=female_stats_df).fit()
    print("\n\nFEMALE MICE NETWORK {} ANCOVA ".format(net) + formula + "\n\n")
    print(model.summary())
    
    #Complete Model
    
    full_df = {}
    full_df["mouse"] = []
    full_df["geno"] = []
    full_df["sex"] = []
    full_df["time"] = []
    full_df["scores"] = []

    for idx, mouse in enumerate(ca_data.mouse.values):
        sex = df_sex[df_sex[0]==mouse][2]
        geno = df_sex[df_sex[0]==mouse][1]
        full_df["mouse"].append([idx+1]*300)
        full_df["sex"].append([sex=="F"]*300)
        full_df["geno"].append([geno=="WT"]*300)
        full_df["time"].append(np.arange(1,301))
        full_df["scores"].append(ca_data[ca_data.mouse==mouse][np.arange(1,301).astype(str)])

    full_df["mouse"] = np.array(full_df["mouse"]).reshape(-1)
    full_df["geno"] = np.array(full_df["geno"]).reshape(-1)
    full_df["sex"] = np.array(full_df["sex"]).reshape(-1)
    full_df["time"] = np.array(full_df["time"]).reshape(-1)
    full_df["scores"] = np.array(full_df["scores"]).reshape(-1)

    df_stats_full = pd.DataFrame.from_dict(full_df)
    
    #formula = "scores ~ geno + time"
    #model = ols(formula,data=df_stats_full).fit()
    #print("\n\nJOINT MODEL MICE NETWORK {} ANCOVA ".format(net) + formula + "\n\n")
    #print(model.summary())
    
    ## With random intercept
    df_stats_full_noNans = df_stats_full[df_stats_full.scores.notnull()].copy()
    formula = "scores ~ geno + sex + time"
    model = smf.mixedlm(formula,data=df_stats_full_noNans,groups=df_stats_full_noNans["mouse"]).fit()
    print("\n\nMixed Effects MODEL MICE NETWORK {} ANCOVA ".format(net) + formula + "\n\n")
    print(model.summary())
```
