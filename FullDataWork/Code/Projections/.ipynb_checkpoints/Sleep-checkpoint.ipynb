```python
import numpy as np
import pandas as pd
import torch
import pickle
import matplotlib.pyplot as plt
from lpne.models import DcsfaNmf
import os, sys

MODEL_FILE = "/hpc/home/mk423/Anxiety/FullDataWork/Models/Final_mt_Model_500_epochs.pt"
DATA_PATH = "/work/mk423/DiazSaline_Hour1/DiazSaline1hr.mat"
PROJECT_PATH = "/hpc/home/mk423/Anxiety/FullDataWork/Projections/"
RUNSHEET_PATH = "/work/mk423/DiazSaline_Hour1/RunSheet_DiazSaline_Hour1.xlsx"
FIGURE_PATH = "/hpc/home/mk423/Anxiety/FullDataWork/Figures/"
LPNE_PATH = "/work/mk423/lpne-data-analysis/"

model = torch.load(MODEL_FILE,map_location="cpu")
model.device="cpu"

sys.path.append(LPNE_PATH)

import data_tools
```

```python
df_runsheet = pd.read_excel(RUNSHEET_PATH)
X_psd, X_coh, X_gc, labels = data_tools.load_data(DATA_PATH,
                                                  feature_list = ["power","coherence","granger"])
```

```python
df_runsheet.head()

```

```python
drug_list = []
for mouse,expdate in zip(labels['windows']['mouse'],labels['windows']['expDate']):
    drug_list.append(df_runsheet[np.logical_and(df_runsheet.MouseID == int(mouse[-5:]), df_runsheet.Date==int(expdate))]["Drug"].values[0])
    #break
```

```python
np.unique(drug_list)
```

```python
s = model.project(np.hstack([X_psd*10,X_coh,X_gc]))
coeffs = np.abs(model.classifier[0].weight[0].detach().cpu().numpy())
agg_score = s[:,:3] @ coeffs

saveDict = {
    "net 1":s[:,0],
    "net 2":s[:,1],
    "net 3":s[:,2],
    "agg score":agg_score,
    "time":labels['windows']['time'],
    "drug":drug_list,
    "mouse":labels['windows']['mouse'],
    "expDate":labels['windows']['expDate'],
}

df = pd.DataFrame.from_dict(saveDict)
df.to_csv(PROJECT_PATH + "Sleep_DZP_Anxiety_Projection.csv")
```

```python
for net in ["net 1","net 2","net 3"]:
    diaz_scores = []
    sal_scores = []
    for drug in ["Diaz","Saline"]:
        drug_mask = df.drug==drug
        for mouse in np.unique(df.mouse):
            mouse_mask = df.mouse==mouse
            mouse_drug_mask = np.logical_and(drug_mask,mouse_mask)
            
            if np.sum(mouse_drug_mask)!=0:
                full_time = np.arange(3600)*np.nan

                mouse_time = df[mouse_drug_mask==1].time
                full_time[mouse_time-1] = df[mouse_drug_mask==1][net]

                if drug=="Diaz":
                    diaz_scores.append(full_time)
                else:
                    sal_scores.append(full_time)
            
    diaz_scores = np.array(diaz_scores)
    sal_scores = np.array(sal_scores)
    
    mean_drug_scores = np.nanmean(diaz_scores,axis=0)
    mean_sal_scores = np.nanmean(sal_scores,axis=0)
    
    sterr_drug_scores = np.nanstd(diaz_scores,axis=0) / np.sqrt(diaz_scores.shape[0])
    sterr_sal_scores = np.nanstd(sal_scores,axis=0) / np.sqrt(sal_scores.shape[0])
    
    x = np.arange(3600)
    
    plt.figure(figsize=(20,10))
    plt.subplot(1,2,1)
    plt.plot(x,mean_drug_scores,label="Diaz (n={})".format(diaz_scores.shape[0]))
    plt.fill_between(x,mean_drug_scores + sterr_drug_scores,mean_drug_scores - sterr_drug_scores,alpha=0.5)
    
    plt.plot(x,mean_sal_scores,label="Sal (n={})".format(sal_scores.shape[0]))
    plt.fill_between(x,mean_sal_scores + sterr_sal_scores,mean_sal_scores - sterr_sal_scores,alpha=0.5)
    
    plt.xlabel("Time (seconds)")
    plt.ylabel(net + " Activity")
    plt.title("1 hr sleep Diaz vs. Saline (1s resolution)")
    plt.legend()
    

    plt.subplot(1,2,2)
    
    minute_mean_drug_scores = np.nanmean(np.nanmean(diaz_scores.reshape(diaz_scores.shape[0],-1,60),axis=2),axis=0)
    minute_mean_sal_scores = np.nanmean(np.nanmean(sal_scores.reshape(sal_scores.shape[0],-1,60),axis=2),axis=0)
    
    minute_sterr_drug_scores = np.nanstd(np.nanmean(diaz_scores.reshape(diaz_scores.shape[0],-1,60),axis=2),axis=0) / np.sqrt(diaz_scores.shape[0])
    minute_sterr_sal_scores = np.nanstd(np.nanmean(sal_scores.reshape(sal_scores.shape[0],-1,60),axis=2),axis=0) / np.sqrt(sal_scores.shape[0])
    
    x=np.arange(60)
    plt.plot(x,minute_mean_drug_scores,label="Diaz (n={})".format(diaz_scores.shape[0]))
    plt.fill_between(x,minute_mean_drug_scores + minute_sterr_drug_scores,minute_mean_drug_scores - minute_sterr_drug_scores,alpha=0.5)
    
    plt.plot(x,minute_mean_sal_scores,label="Sal (n={})".format(sal_scores.shape[0]))
    plt.fill_between(x,minute_mean_sal_scores + minute_sterr_sal_scores,minute_mean_sal_scores - minute_sterr_sal_scores,alpha=0.5)
    
    plt.xlabel("Time (minutes)")
    plt.ylabel(net + " Activity")
    plt.title("1 hr sleep Diaz vs. Saline (minute averaged)")
    plt.legend()
    
    
    plt.show()
```
