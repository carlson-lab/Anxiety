```python
import numpy as np
import pandas as pd
import torch
import pickle
import matplotlib.pyplot as plt
from lpne.models import DcsfaNmf
from sklearn.metrics import roc_auc_score

MODEL_FILE = "/hpc/home/mk423/Anxiety/FullDataWork/Models/Final_mt_Model_500_epochs.pt"
DATA_PATH = "/work/mk423/Anxiety/"
PROJECT_PATH = "/hpc/home/mk423/Anxiety/FullDataWork/Projections/"
data_file = DATA_PATH + "final_FLX_test.pkl"
proj_file = PROJECT_PATH + "FLX_Holdout_Projections.csv"
mean_file = PROJECT_PATH + "{}_FLX_Holdout_mean_scores.csv"

model = torch.load(MODEL_FILE,map_location="cpu")
model.device="cpu"

import os, sys
umc_data_tools_path = "/hpc/home/mk423/Anxiety/Universal-Mouse-Code/"
sys.path.append(umc_data_tools_path)
import umc_data_tools as umc_dt

feature_list = ["X_psd","X_coh","X_gc"]
feature_weights = [10,1,1]

def get_3_net_aucs(s,y,y_group=None):
    
    auc_mean_list = []
    auc_stderr_list =[]
    
    if y_group is None:
        y_group = np.ones(s.shape[0])
        
    for i in range(3):
        auc_dict = umc_dt.lpne_auc(y,y,y_group,s[:,i].reshape(-1,1),mannWhitneyU=True)
        mean = np.mean([auc_dict[key][0] for key in auc_dict.keys() if key != "auc_method"])
        stderr = np.std([auc_dict[key][0] for key in auc_dict.keys() if key != "auc_method"]) / np.sqrt(len(auc_dict.keys()))
        
        auc_mean_list.append(mean)
        auc_stderr_list.append(stderr)
        
    return auc_mean_list, auc_stderr_list

def get_3_net_auc_dict(s,y,y_group=None):
    auc_dict_list = []
    
    if y_group is None:
        y_group = np.ones(s.shape[0])
        
    for i in range(3):
        auc_dict = umc_dt.lpne_auc(y,y,y_group,s[:,i].reshape(-1,1),mannWhitneyU=True)
        auc_dict_list.append(auc_dict)
        
    return auc_dict_list
```

```python
from scipy.stats import mannwhitneyu

def mw_auc(s,y):
    pos_mask = y==1
    neg_mask = y==0
    
    U, pval = mannwhitneyu(s[pos_mask==1],s[neg_mask==1])
    
    auc = U/(np.sum(pos_mask)*np.sum(neg_mask))
    
    return auc

results_dict_comp = {}
results_dict_s1 = {}
results_dict_s2 = {}
results_dict_s3 = {}

for mouse in np.unique(df["mouse"]):
    time_mouse_mask = np.logical_and(df["mouse"]==mouse,df["time"]>(30*60))
    y = df[time_mouse_mask]["flx"].to_numpy()
    y_minute = y[:-(y.shape[0]%60)].reshape(-1,60)
    y_minute = np.mean(y_minute,axis=1) > 0.5
    
    s_comp = df[time_mouse_mask]["comp score"].to_numpy()
    s_1 = df[time_mouse_mask]["net 1 scores"].to_numpy()
    s_2 = df[time_mouse_mask]["net 2 scores"].to_numpy()
    s_3 = df[time_mouse_mask]["net 3 scores"].to_numpy()
    
    s_comp_minute = s_comp[:-(s_comp.shape[0]%60)].reshape(-1,60)
    s_comp_minute = np.mean(s_comp_minute,axis=1)
    s_1_minute = s_1[:-(s_1.shape[0]%60)].reshape(-1,60)
    s_1_minute = np.mean(s_1_minute,axis=1)
    s_2_minute = s_2[:-(s_2.shape[0]%60)].reshape(-1,60)
    s_2_minute = np.mean(s_2_minute,axis=1)
    s_3_minute = s_3[:-(s_3.shape[0]%60)].reshape(-1,60)
    s_3_minute = np.mean(s_3_minute,axis=1)
    
    comp_net_auc = mw_auc(s_comp_minute,y_minute)
    s_1_auc = mw_auc(s_1_minute,y_minute)
    s_2_auc = mw_auc(s_2_minute,y_minute)
    s_3_auc = mw_auc(s_3_minute,y_minute)
    
    
    results_dict_comp[mouse] = comp_net_auc
    results_dict_s1[mouse] = s_1_auc
    results_dict_s2[mouse] = s_2_auc
    results_dict_s3[mouse] = s_3_auc
    print("{} AUCs comp score: {:.2f} net 1: {:.2f}, net 2: {:.2f}, net 3: {:.2f}".format(mouse,comp_net_auc,s_1_auc,s_2_auc,s_3_auc))
```

```python
max_minute = 61
num_seconds = 300
min_df = {}

min_df["5_minute_net_1_scores"] = []
min_df["5_minute_net_2_scores"] = []
min_df["5_minute_time"] = []
min_df["5_minute_mouse"] = []
min_df["5_minute_flx"] = []
min_df["5_minute_expDate"] = []

for mouse in np.unique(df["mouse"]):
    mouse_mask = df["mouse"]==mouse
    mouse_expdates = np.unique(df[mouse_mask]["expDate"])
    
    for expDate in mouse_expdates:
        print(mouse,expDate)
        expDate_mask = df["expDate"]==expDate
        
        for rec_flx in [0,1]:
            
            flx_mask = df["flx"]==rec_flx
            
            for i in range(int(max_minute*60/num_seconds)):
                start_time = i*num_seconds
                end_time = (i+1)*num_seconds

                time_mask = np.logical_and(df["time"] > start_time, df["time"] <= end_time)

                mask = np.logical_and(mouse_mask,expDate_mask)
                mask = np.logical_and(mask,time_mask)
                mask = np.logical_and(mask,flx_mask)
                
                
                if np.sum(mask) > 0:
                    flx = df[mask]["flx"].values[0]
                    min_df["5_minute_net_1_scores"].append(np.mean(df[mask]["net 1 scores"]))
                    min_df["5_minute_net_2_scores"].append(np.mean(df[mask]["net 2 scores"]))
                    min_df["5_minute_time"].append(start_time)
                    min_df["5_minute_mouse"].append(mouse)
                    min_df["5_minute_expDate"].append(expDate)
                    min_df["5_minute_flx"].append(flx)

min_df = pd.DataFrame.from_dict(min_df)
min_df.to_csv(PROJECT_PATH + "FLX_Net_1_and_2_5_Minute_Projections.csv",index=False)
```
