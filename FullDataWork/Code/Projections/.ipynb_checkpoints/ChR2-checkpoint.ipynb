```python
import numpy as np
import pandas as pd
import torch
import pickle
import matplotlib.pyplot as plt
from lpne.models import DcsfaNmf

MODEL_FILE = "/hpc/home/mk423/Anxiety/FullDataWork/Models/Final_mt_Model_500_epochs.pt"
DATA_PATH = "/work/mk423/Anxiety/"
PROJECT_PATH = "/hpc/home/mk423/Anxiety/FullDataWork/Projections/"
FIGURE_PATH = "/hpc/home/mk423/Anxiety/FullDataWork/Figures/"
chr2_file = DATA_PATH + "ChR2_pickle_file.pkl"
proj_file = PROJECT_PATH + "chr2_projection.csv"
mean_file = PROJECT_PATH + "chr2_mean_scores.csv"
model = torch.load(MODEL_FILE,map_location="cpu")
model.device="cpu"


FEATURE_LIST = ['X_psd','X_coh','X_gc']
FEATURE_WEIGHT = [10,1,1]

import os, sys
umc_data_tools_path = "/hpc/home/mk423/Anxiety/Universal-Mouse-Code/"
sys.path.append(umc_data_tools_path)
import umc_data_tools as umc_dt

def get_3_net_aucs(s,y,y_group=None):
    
    auc_mean_list = []
    auc_stderr_list =[]
    
    if y_group is None:
        y_group = np.ones(s.shape[0])
        
    for i in range(3):
        auc_dict = umc_dt.lpne_auc(y,y,y_group,s[:,i].reshape(-1,1),mannWhitneyU=True)
        mean = np.mean([auc_dict[key][0] for key in auc_dict.keys() if key != "auc_method"])
        stderr = np.std([auc_dict[key][0] for key in auc_dict.keys() if key != "auc_method"]) / np.sqrt(len(auc_dict.keys()))
        
        auc_mean_list.append(mean)
        auc_stderr_list.append(stderr)
        
    return auc_mean_list, auc_stderr_list

def get_3_net_auc_dict(s,y,y_group=None):
    auc_dict_list = []
    
    if y_group is None:
        y_group = np.ones(s.shape[0])
        
    for i in range(3):
        auc_dict = umc_dt.lpne_auc(y,y,y_group,s[:,i].reshape(-1,1),mannWhitneyU=True)
        auc_dict_list.append(auc_dict)
        
    return auc_dict_list
```

```python
dataDict = pickle.load(open(chr2_file,"rb"))

X = np.hstack([weight*dataDict[feature] for weight,feature in zip(FEATURE_WEIGHT,FEATURE_LIST)])
y_time = dataDict['y_time']
y_mouse = np.array(dataDict['y_mouse'])
y_expDate = dataDict['y_expDate']
y_BLaser = np.array(dataDict['y_BLaser'])
y_YLaser = np.array(dataDict['y_YLaser'])
y_Laser = np.logical_or(y_BLaser,y_YLaser)
```

```python
blaser_auc = model.score(X[y_Laser==1],
                         y_BLaser[y_Laser==1],
                         y_mouse[y_Laser==1],
                         return_dict=True)

blaser_auc_list = [blaser_auc[key] for key in blaser_auc.keys() if not np.isnan(blaser_auc[key])]

blaser_mean = np.mean(blaser_auc_list)
blaser_sterr = np.std(blaser_auc_list) / np.sqrt(len(blaser_auc_list))

print("Blue Laser vs Yellow Laser AUC {:.3f} +/- {:.3f} (n={})".format(blaser_mean,blaser_sterr,len(blaser_auc_list)))
```

```python
s = model.project(X)
blaser_auc_3_net = get_3_net_auc_dict(s[y_Laser==1],y_BLaser[y_Laser==1],y_mouse[y_Laser==1])

y_time_aw_list = []
y_mouse_aw_list = []
y_expDate_aw_list = []
y_YLaser_aw_list = []
y_BLaser_aw_list = []
y_Laser_aw_list = []

s_aw_list = []
for mouse in np.unique(y_mouse):
    for expDate in np.unique(y_expDate):
        expDate_mask = np.array(y_expDate)==expDate
        mouse_mask = y_mouse==mouse
        mask = np.logical_and(expDate_mask,mouse_mask)
        
        if np.sum(mask) >= 1:
            
            temp_time = np.array(y_time)[mask==1]
            temp_time_aw = np.arange(np.max(temp_time)+1)
            
            temp_y_YLaser = np.zeros(temp_time_aw.shape)*np.nan
            temp_y_YLaser[temp_time - 1] = y_YLaser[mask==1]
            
            temp_y_BLaser = np.zeros(temp_time_aw.shape)*np.nan
            temp_y_BLaser[temp_time - 1] = y_BLaser[mask==1]
            
            temp_y_Laser = np.zeros(temp_time_aw.shape)*np.nan
            temp_y_Laser[temp_time - 1] = y_Laser[mask==1]
            
            temp_s = np.zeros((temp_time_aw.shape[0],s.shape[1]))*np.nan
            temp_s[temp_time - 1] = s[mask==1]
            
            temp_y_mouse = np.array([mouse for _ in temp_time_aw])
            temp_y_expDate = np.array([expDate for _ in temp_time_aw])
            
            
            
            y_time_aw_list.append(temp_time_aw)
            y_mouse_aw_list.append(temp_y_mouse)
            y_expDate_aw_list.append(temp_y_expDate)
            y_YLaser_aw_list.append(temp_y_YLaser)
            y_BLaser_aw_list.append(temp_y_BLaser)
            y_Laser_aw_list.append(temp_y_Laser)
            s_aw_list.append(temp_s)
            
y_time_aw = np.hstack(y_time_aw_list)
y_mouse_aw = np.hstack(y_mouse_aw_list)
y_expDate_aw = np.hstack(y_expDate_aw_list)
y_YLaser_aw = np.hstack(y_YLaser_aw_list)
y_BLaser_aw = np.hstack(y_BLaser_aw_list)
y_Laser_aw = np.hstack(y_Laser_aw_list)
s_aw = np.vstack(s_aw_list)
```

```python
coeffs = np.abs(model.classifier[0].weight[0].detach().cpu().numpy())

results_dict = {
    "composite BLaser vs YLaser AUC":[blaser_auc[mouse][0] for mouse in y_mouse_aw],
    "net 1 BLaser vs YLaser AUC":[blaser_auc_3_net[0][mouse][0] for mouse in y_mouse_aw],
    "net 2 BLaser vs YLaser AUC":[blaser_auc_3_net[1][mouse][0] for mouse in y_mouse_aw],
    "net 3 BLaser vs YLaser AUC":[blaser_auc_3_net[2][mouse][0] for mouse in y_mouse_aw],
    "composite scores":s_aw[:,:3]@coeffs,
    "net 1 scores":s_aw[:,0],
    "net 2 scores":s_aw[:,1],
    "net 3 scores":s_aw[:,2],
    "y_BLaser":y_BLaser_aw,
    "y_YLaser":y_YLaser_aw,
    "y_Laser":y_Laser_aw,
    "mouse":y_mouse_aw,
    "time":y_time_aw,
    "expDate":y_expDate_aw,
}

df = pd.DataFrame.from_dict(results_dict)
df.to_csv(proj_file)
df.head()
```

```python
#Get per mouse average scores, average scores in HC and average scores in EPM
mouse_list = []
avg_laser_list = []
avg_BLaser_list = []
avg_YLaser_list = []

for mouse in np.unique(y_mouse):
    mouse_mask = y_mouse==mouse
    mouse_and_laser = np.logical_and(mouse_mask,y_Laser)
    mouse_and_BLaser = np.logical_and(mouse_mask,y_BLaser)
    mouse_and_YLaser = np.logical_and(mouse_mask,y_YLaser)
    
    avg_laser_score = np.mean(s[mouse_and_laser==1,:3],axis=0)
    avg_BLaser_score = np.mean(s[mouse_and_BLaser==1,:3],axis=0)
    avg_YLaser_score = np.mean(s[mouse_and_YLaser==1,:3],axis=0)
    
    mouse_list.append(mouse)
    avg_laser_list.append(avg_laser_score)
    avg_BLaser_list.append(avg_BLaser_score)
    avg_YLaser_list.append(avg_YLaser_score)
    
avg_laser_list = np.array(avg_laser_list)
avg_BLaser_list = np.array(avg_BLaser_list)
avg_YLaser_list = np.array(avg_YLaser_list)

coeffs = np.abs(model.classifier[0].weight[0].detach().cpu().numpy())

mag_laser = avg_laser_list*coeffs
mag_blaser = avg_BLaser_list*coeffs
mag_ylaser = avg_YLaser_list*coeffs

net_impact_laser = mag_laser / np.sum(mag_laser,axis=1).reshape(-1,1)
net_impact_blaser = mag_blaser / np.sum(mag_blaser,axis=1).reshape(-1,1)
net_impact_ylaser = mag_ylaser / np.sum(mag_ylaser,axis=1).reshape(-1,1)


proj_dict = {
    "mouse":mouse_list,
    
    "composite BLaser vs YLaser AUC":[blaser_auc[mouse][0] for mouse in mouse_list],
    "net 1 BLaser vs YLaser AUC":[blaser_auc_3_net[0][mouse][0] for mouse in mouse_list],
    "net 2 BLaser vs YLaser AUC":[blaser_auc_3_net[1][mouse][0] for mouse in mouse_list],
    "net 3 BLaser vs YLaser AUC":[blaser_auc_3_net[2][mouse][0] for mouse in mouse_list],
    
    "composite avgLaserScore":np.sum(mag_laser,axis=1),
    "composite avgBLaserScore":np.sum(mag_blaser,axis=1),
    "composite avgYLaserScore":np.sum(mag_ylaser,axis=1),
    
    "net 1 avgLaserScore":avg_laser_list[:,0],
    "net 1 avgBLaserScore":avg_BLaser_list[:,0],
    "net 1 avgYLaserScore":avg_YLaser_list[:,0],
    "net 2 avgLaserScore":avg_laser_list[:,1],
    "net 2 avgBLaserScore":avg_BLaser_list[:,1],
    "net 2 avgYLaserScore":avg_YLaser_list[:,1],
    "net 3 avgLaserScore":avg_laser_list[:,2],
    "net 3 avgBLaserScore":avg_BLaser_list[:,2],
    "net 3 avgYLaserScore":avg_YLaser_list[:,2],
    
    "net 1 avgLaserImpact":net_impact_laser[:,0],
    "net 1 avgBLaserImpact":net_impact_blaser[:,0],
    "net 1 avgYLaserImpact":net_impact_ylaser[:,0],
    "net 2 avgLaserImpact":net_impact_laser[:,1],
    "net 2 avgBLaserImpact":net_impact_blaser[:,1],
    "net 2 avgYLaserImpact":net_impact_ylaser[:,1],
    "net 3 avgLaserImpact":net_impact_laser[:,2],
    "net 3 avgBLaserImpact":net_impact_blaser[:,2],
    "net 3 avgYLaserImpact":net_impact_ylaser[:,2],
}
df_means = pd.DataFrame.from_dict(proj_dict)

df_means.to_csv(mean_file)
df_means.head()
```
