```python
import numpy as np
import pandas as pd
import torch
import pickle
import matplotlib.pyplot as plt
from lpne.models import DcsfaNmf
import os, sys

MODEL_FILE = "/hpc/home/mk423/Anxiety/FullDataWork/Models/Final_mt_Model_500_epochs.pt"
DATA_PATH = "/work/mk423/Anxiety/FearConditioning_Data.pkl"
PROJECT_PATH = "/hpc/home/mk423/Anxiety/FullDataWork/Projections/"
FIGURE_PATH = "/hpc/home/mk423/Anxiety/FullDataWork/Figures/"
LPNE_PATH = "/work/mk423/lpne-data-analysis"
sys.path.append(LPNE_PATH)

import data_tools


model = torch.load(MODEL_FILE,map_location="cpu")
model.device="cpu"

feature_list = ["X_psd","X_coh","X_gc"]
feature_weights = [10,1,1]

group_1_mice = [
"Mouse04191",
"Mouse04201",
"Mouse04203",
"Mouse04215",
"Mouse3203",
"Mouse3223",
"Mouse39114",
"Mouse39121",
"Mouse39124",
"Mouse39133",
"Mouse9001",
"Mouse9011",
"Mouse9013",
"Mouse9022",
"Mouse9071",
"Mouse9072",
"Mouse9084",
]

group_2_mice = [
"Mouse04193",
"Mouse04202",
"Mouse04211",
"Mouse04213",
"Mouse3205",
"Mouse3235",
"Mouse3252",
"Mouse39122",
"Mouse39125",
"Mouse39132",
"Mouse39135",
"Mouse9002",
"Mouse9003",
"Mouse9021",
"Mouse9065",
"Mouse9074",
]
```

```python
X_psd, X_coh, X_gc, labels = data_tools.load_data("/work/mk423/Anxiety/Tone_Start_FC.mat",feature_list = ["power","coherence","granger"])

X = np.hstack([X_psd*10,X_coh,X_gc])
y_mouse = np.array(labels['windows']['mouse'])
y_time = np.array(labels['windows']['time'])
y_frames = np.array(labels['windows']['frame'])

s = model.project(X)
coeffs = model.classifier[0].weight[0].detach().cpu().numpy()
s_agg = s[:,:3] @ coeffs
```

```python
start_tone_dict = {}
for tone_number in range(7):
    start_tone_dict["Tone {}".format(tone_number + 1)] = {}
    for mouse in np.unique(y_mouse):
        temp_scores = np.zeros((50,4))*np.nan
        
        mouse_mask = y_mouse==mouse
        unique_frames = np.unique(y_frames[mouse_mask==1])
        if unique_frames.shape[0] == 7:
            tone_frame = unique_frames[tone_number]
            frame_mask = y_frames == tone_frame

            mask = np.logical_and(mouse_mask,frame_mask)

            sliced_time = y_time[mask==1] - np.min(y_time[mask==1])

            temp_scores[sliced_time] = np.hstack([s[mask==1,:3],s_agg[mask==1].reshape(-1,1)])

            start_tone_dict["Tone {}".format(tone_number+1)][mouse] = temp_scores
        
for tone_number in range(7):
    g1_segment_list = []
    g2_segment_list = []
    for mouse in np.unique(y_mouse):
        if mouse in group_1_mice:
            g1_segment_list.append(start_tone_dict["Tone {}".format(tone_number+1)][mouse])
        if mouse in group_2_mice:
            g2_segment_list.append(start_tone_dict["Tone {}".format(tone_number+1)][mouse])

    g1_mean = np.nanmean(np.array(g1_segment_list),axis=0)
    g1_stderr = np.nanstd(np.array(g1_segment_list),axis=0) / np.sqrt(np.array(g1_segment_list).shape[0])

    g2_mean = np.nanmean(np.array(g2_segment_list),axis=0)
    g2_stderr = np.nanstd(np.array(g2_segment_list),axis=0) / np.sqrt(np.array(g2_segment_list).shape[0])

    plt.figure(figsize=(30,4))
    ranges = [(0.04,0.15),(0.08,0.25),(0.01,0.1),(2,5)]
    for net in range(4):
        plt.subplot(1,4,net+1)
        plt.plot(g1_mean[:,net],label="group 1")
        plt.fill_between(np.arange(50), g1_mean[:,net] + g1_stderr[:,net],g1_mean[:,net] - g1_stderr[:,net],alpha=0.5)

        plt.plot(g2_mean[:,net],label="group 2")
        plt.fill_between(np.arange(50), g2_mean[:,net] + g2_stderr[:,net],g2_mean[:,net] - g2_stderr[:,net],alpha=0.5)

        plt.axvline(10,linestyle="dashed",color="red")
        plt.xticks(np.arange(0,50,10), np.arange(0,50,10) - 10)
        plt.ylim([ranges[net][0],ranges[net][1]])
        
        if net < 3:
            plt.title("Start Tone Event: {} Net: {}".format(tone_number,net))
        else:
            plt.title("Start Tone Event: {} Net: {}".format(tone_number,"Aggregate"))
        plt.legend()
    plt.suptitle("Event: {}".format(tone_number))        
    plt.show()

```
